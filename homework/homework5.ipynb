W5A4 VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên.

```python
def la_so_hoan_hao(n):
    if n <= 1:
        return False
    
    tong_cac_uoc = 0
    for i in range(1, (n // 2) + 1):
        if n % i == 0:
            tong_cac_uoc += i
            
    return tong_cac_uoc == n

n = int(input("Nhập số cần kiểm tra: "))
if la_so_hoan_hao(n):
    print(f"{n} là số hoàn hảo")
else:
    print(f"{n} không phải là số hoàn hảo")
```

W5A12
Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và
0 <= |r| < |divisor|.

```python
def divide_with_remainder(dividend, divisor):
    if divisor == 0:
        return "Lỗi: Không thể chia cho 0"
    
    q = dividend // divisor
    r = dividend % divisor
    
    return (q, r)

a,b=map(int,input().split())
print(divide_with_remainder(a,b))
```

W5A18 VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.


```python
def la_so_doi_xuong(n):
    if n < 0:
        return False
    
    so_goc = n
    so_dao_nguoc = 0
    
    while n > 0:
        chu_so = n % 10
        so_dao_nguoc = so_dao_nguoc * 10 + chu_so
        n //= 10 
        
    return so_goc == so_dao_nguoc
n=int(input())

print(la_so_doi_xuong(n))
```


W5A23 Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0.

```python
import math
a = float(input("nhap a: "))
b = float(input("nhap b: "))
c = float(input("nhap c: "))
def quadratic_roots_count(a, b, c):
    delta=b**2-4*a*c
    if delta <0:
        return "Phương trình vô nghiệm"
    elif delta == 0 :
        x=-b/(2*a)
        print("Nghiệm là:",x)
        return x
    else:
        x1=((-b-math.sqrt(delta))/(2*a))
        x2=((-b+math.sqrt(delta))/(2*a))
        return x1,x2
print(quadratic_roots_count(a,b,c))
```

W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).


```python
import math
A=list(map(float,input().split()))
B=list(map(float,input().split()))
C=list(map(float,input().split()))

# A_B=(A[0]-B[0],A[1]-B[1])
ab=math.sqrt((A[0]-B[0])**2+(A[1]-B[1])**2)    
# C_B=(C[0]-B[0],C[1]-B[1])
bc=math.sqrt((C[0]-B[0])**2+(C[1]-B[1])**2) 
ac=math.sqrt((A[0]-C[0])**2+(A[1]-C[1])**2) 
so=min((ac**2-ab**2-bc**2)/(-2*ab*bc),1)
goc=math.acos(so)
print(math.degrees(goc))
```
